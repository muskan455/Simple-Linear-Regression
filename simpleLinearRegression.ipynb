**Question 1:  What is Simple Linear Regression?**

ans = Simple linear regression is a statistical method used to model the relationship between two variables: one independent variable (predictor) and one dependent variable (response). It assumes a linear relationship, meaning the change in the dependent variable is proportional to the change in the independent variable. The goal is to find the best-fitting straight line (regression line) that predicts the dependent variable based on the independent variable.
The equation of the line is:

$ y = \beta_0 + \beta_1x + \epsilon $

where:

$ y $: Dependent variable (what you’re predicting)

$ x $: Independent variable (what you’re using to predict)

$ \beta_0 $: Intercept (value of $ y $ when $ x = 0 $)

$ \beta_1 $: Slope (change in $ y $ for a unit change in $ x $)

$ \epsilon $: Error term (random variation not explained by the model)

Example : predict the price of house based on the number of rooms and price of area


**Question 2: What are the key assumptions of Simple Linear Regression?**

ans = The key assumptions of simple linear regression are:

**Linearity:** The relationship between the independent variable ($ x $) and the dependent variable ($ y $) is linear, meaning $ y $ changes proportionally with $ x $.
**Independence:** Observations are independent of each other. The value of the dependent variable for one observation does not influence another.
**Homoscedasticity:** The variance of the error terms ($ \epsilon $) is constant across all values of the independent variable. This means the spread of residuals is roughly the same for all predicted values.

**Normality of Errors:** The error terms are normally distributed with a mean of zero. This assumption is particularly important for hypothesis testing and confidence intervals.

**No Perfect Multicollinearity:** Since simple linear regression involves only one independent variable, this assumption is inherently satisfied (but relevant in multiple regression).

**No Significant Outliers:** The model assumes that there are no extreme outliers that disproportionately influence the regression line.

**Question 3: What is heteroscedasticity, and why is it important to address in regression
models?**

ans = **Heteroscedasticity** refers to the condition in which the variance of the error terms (residuals) in a regression model is not constant across all levels of the independent variable(s). In other words, the spread or variability of the residuals changes as the values of the independent variable (\( x \)) change. This violates the homoscedasticity assumption of simple linear regression, which requires constant variance of errors.

### Characteristics of Heteroscedasticity:
- **Visual Representation**: In a residual plot (residuals vs. predicted values), heteroscedasticity often appears as a funnel shape, where residuals spread out more at certain levels of the predicted values.
- **Example**: In predicting house prices based on size, the variability in price prediction errors might increase for larger houses, showing greater spread in residuals for higher values of size.

### Why It’s Important to Address in Regression Models:
1. **Biased Standard Errors**: Heteroscedasticity leads to unreliable standard errors for the regression coefficients. This affects the accuracy of hypothesis tests (e.g., t-tests for coefficient significance) and confidence intervals, potentially leading to incorrect conclusions about the significance of predictors.

2. **Inefficient Estimates**: While the regression coefficients remain unbiased, they are no longer the most efficient (i.e., they don’t have the smallest possible variance). This reduces the precision of the model.

3. **Invalid Inference**: Many statistical tests in regression (e.g., F-tests, t-tests) assume homoscedasticity. Heteroscedasticity can lead to misleading p-values, causing you to incorrectly reject or fail to reject null hypotheses.

4. **Poor Predictive Performance**: Models with heteroscedasticity may perform poorly in certain regions of the data where variance is high, leading to less reliable predictions.

### Addressing Heteroscedasticity:
- **Transformations**: Apply transformations to the dependent variable (e.g., log or square root) to stabilize variance.
- **Weighted Least Squares (WLS)**: Use a regression method that assigns weights to observations based on the inverse of their variance, giving less influence to observations with higher variability.
- **Robust Standard Errors**: Use robust standard errors (e.g., Huber-White standard errors) to account for heteroscedasticity without altering the model structure.
- **Model Respecification**: Include additional predictors or use a different model (e.g., nonlinear regression) if the heteroscedasticity stems from a misspecified relationship.

### Detection:
- **Residual Plots**: Plot residuals against predicted values or the independent variable to visually check for patterns like a funnel shape.
- **Statistical Tests**: Use tests like the Breusch-Pagan test or White’s test to formally detect heteroscedasticity.

Addressing heteroscedasticity ensures more reliable statistical inference and improves the overall validity of the regression model.

**Question 4: What is Multiple Linear Regression?**

ans = Multiple Linear Regression is an extension of simple linear regression that models the relationship between one dependent variable (response) and two or more independent variables (predictors). It aims to predict the dependent variable using a linear combination of the independent variables, assuming a linear relationship between them.
Model Equation:
The equation for multiple linear regression is:

$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon $

where:

$ y $: Dependent variable (what you’re predicting)

$ x_1, x_2, \dots, x_n $: Independent variables (predictors)

$ \beta_0 $: Intercept (value of $ y $ when all $ x_i = 0 $)

$ \beta_1, \beta_2, \dots, \beta_n $: Coefficients (slopes) representing the change in $ y $ for a one-unit change in each $ x_i $, holding other variables constant

$ \epsilon $: Error term (random variation not explained by the model)

Example:
To predict house price ($ y $), you might use predictors like size ($ x_1 $), number of bedrooms ($ x_2 $), and age of the house ($ x_3 $). The model estimates how each factor contributes to the price while accounting for the others.

**Question 5: What is polynomial regression, and how does it differ from linear
regression?**

ans = Polynomial Regression is a type of regression analysis that models the relationship between a dependent variable ($ y $) and an independent variable ($ x $) as an $ n $-th degree polynomial. Unlike simple linear regression, which assumes a straight-line relationship, polynomial regression allows for curved relationships by including higher-order terms (e.g., $ x^2, x^3 $) in the model.
Polynomial Regression Equation:

The equation for polynomial regression of degree $ n $ is:
$ y = \beta_0 + \beta_1x + \beta_2x^2 + \dots + \beta_nx^n + \epsilon $

where:

$ y $: Dependent variable

$ x $: Independent variable

$ \beta_0 $: Intercept
$ \beta_1, \beta_2, \dots, \beta_n $: Coefficients for each term
$ x^2, x^3, \dots, x^n $: Higher-order terms of the independent variable
$ \epsilon $: Error term

**How Polynomial Regression Differs from Linear Regression:**

**Relationship Modeled:**

**Linear Regression:** Assumes a linear relationship between $ x $ and $ y $, represented by a straight line ($ y = \beta_0 + \beta_1x + \epsilon $).

**Polynomial Regression:** Models a nonlinear relationship by including polynomial terms, allowing the model to fit curves (e.g., quadratic, cubic).


**Flexibility:**

**Linear Regression:** Limited to straight-line relationships, which may not capture complex patterns in the data.

**Polynomial Regression:** Can model more complex, nonlinear patterns by increasing the degree of the polynomial, making it more flexible for fitting data with curvature.


**Model Complexity:**

**Linear Regression:** Simpler, with only two parameters ($ \beta_0, \beta_1 $).

**Polynomial Regression:** More complex, with additional parameters for each polynomial term ($ \beta_2, \beta_3, \dots $), which increases with the degree of the polynomial.


**Overfitting Risk:**

**Linear Regression:** Less prone to overfitting but may underfit if the true relationship is nonlinear.

**Polynomial Regression:** Higher-degree polynomials can lead to overfitting, where the model fits the noise in the data rather than the underlying trend, especially with small datasets.


**Interpretability:**

**Linear Regression:** Coefficients are straightforward to interpret (change in $ y $ per unit change in $ x $).

**Polynomial Regression:** Coefficients for higher-order terms (e.g., $ x^2 $) are less intuitive, as they describe the effect of squared or higher powers of $ x $.


**Computation:**

**Linear Regression:** Computationally simpler due to fewer parameters.

**Polynomial Regression:** Requires more computational effort, especially for higher-degree polynomials, and may involve transforming $ x $ into polynomial terms.



**Example:**

**Linear Regression:** Predicting house price based on size, assuming price increases linearly with size.

**Polynomial Regression:** Predicting house price where the relationship is curved (e.g., price increases rapidly for small sizes but levels off for larger sizes), modeled as $ y = \beta_0 + \beta_1x + \beta_2x^2 + \epsilon $.

**Question 6:  Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Fit simple linear regression model
model = LinearRegression()
model.fit(X, Y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Predict values for the regression line
X_line = np.array([[0], [6]])  # Extend slightly beyond data for visualization
Y_line = model.predict(X_line)

# Plot data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_line, Y_line, color='red', label=f'Regression Line: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Print model parameters
print(f"Slope (β1): {slope:.2f}")
print(f"Intercept (β0): {intercept:.2f}")
print(f"R-squared: {model.score(X, Y):.4f}")
```

**Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Prepare features (X) and target (y)
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit multiple linear regression model
model = LinearRegression()
model.fit(X, y)

# Get coefficients and intercept
intercept = model.intercept_
coefficients = model.coef_
r_squared = model.score(X, y)

# Calculate VIF for each feature
X_vif = X.assign(Intercept=1)  # Add intercept for VIF calculation
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1]-1)]

# Print model results
print("Multiple Linear Regression Results:")
print(f"Intercept (β0): {intercept:.2f}")
print(f"Coefficient for Area (β1): {coefficients[0]:.2f}")
print(f"Coefficient for Rooms (β2): {coefficients[1]:.2f}")
print(f"R-squared: {r_squared:.4f}")
print("\nVariance Inflation Factor (VIF):")
print(vif_data)

# Interpret VIF
print("\nVIF Interpretation:")
for index, row in vif_data.iterrows():
    print(f"{row['Feature']}: VIF = {row['VIF']:.2f}")
    if row['VIF'] > 10:
        print(f"  - High multicollinearity (VIF > 10)")
    elif row['VIF'] > 5:
        print(f"  - Moderate multicollinearity (VIF > 5)")
    else:
        print(f"  - Low or no multicollinearity (VIF ≤ 5)")
```

**Question 8:  Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
3
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial regression model (degree 2)
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
polyreg.fit(X, Y)

# Get coefficients
poly_features = PolynomialFeatures(degree)
X_poly = poly_features.fit_transform(X)
model = LinearRegression().fit(X_poly, Y)
coef = model.coef_
intercept = model.intercept_

# Create points for smooth curve
X_curve = np.linspace(0, 6, 100).reshape(-1, 1)
Y_curve = polyreg.predict(X_curve)

# Plot data points and polynomial curve
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_curve, Y_curve, color='red', label=f'Polynomial Fit (degree {degree})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()

# Print model parameters
print(f"Polynomial Equation: y = {intercept:.2f} + {coef[1]:.2f}x + {coef[2]:.2f}x^2")
print(f"R-squared: {polyreg.score(X, Y):.4f}")
```

**Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Create residuals plot
plt.scatter(Y_pred, residuals, color='blue', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Line')
plt.xlabel('Predicted Y')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.legend()
plt.grid(True)
plt.show()

# Assessment of heteroscedasticity
print("Residuals:", residuals)
print("Assessment: Visually inspect the plot. If residuals show a consistent spread across predicted Y values, variance is likely homoscedastic. A funnel shape or systematic change in spread suggests heteroscedasticity.")
```

**Question 10: Imagine you are a data scientist working for a real estate company. You

**Addressing Heteroscedasticity and Multicollinearity in House Price Prediction Model**

As a data scientist at a real estate company, building a reliable model to predict house prices based on features like area (e.g., square footage), number of rooms, and location (e.g., encoded as categorical dummies or geospatial coordinates) is crucial for tasks like valuation, investment advice, or portfolio optimization. Linear regression is a natural starting point due to its interpretability, but issues like heteroscedasticity (non-constant residual variance) and multicollinearity (high correlation between predictors) can undermine the model's validity, leading to biased inference, inefficient estimates, and poor out-of-sample performance. Heteroscedasticity often arises in real estate data because price variance increases with house size or in high-value locations (e.g., larger homes have more variable prices due to luxury add-ons). Multicollinearity might occur if area and number of rooms are highly correlated (bigger houses tend to have more rooms).

Below, I'll outline a systematic, step-by-step approach to diagnose and address these issues, followed by broader steps to ensure model robustness. This process assumes we're using Python libraries like scikit-learn, statsmodels, or pandas for implementation, and we'd iterate based on diagnostics.

#### Step 1: Diagnose the Issues
Before fixing, confirm the problems empirically to avoid over-correction.
- **For Heteroscedasticity**:
  - Fit an initial ordinary least squares (OLS) model: `price ~ area + rooms + location_dummies`.
  - Plot residuals vs. fitted values (residuals plot) to visually check for patterns like a "funnel" shape (increasing spread at higher predictions).
  - Run statistical tests: Breusch-Pagan test (for variance proportional to fitted values) or White test (for general heteroscedasticity) using statsmodels. If p-value < 0.05, confirm the issue.
- **For Multicollinearity**:
  - Compute the correlation matrix for predictors (e.g., Pearson correlation between area and rooms; threshold > 0.7-0.8 flags high correlation).
  - Calculate Variance Inflation Factors (VIF) for each predictor: VIF = 1 / (1 - R²) from regressing one variable on others. If VIF > 5 (or >10 for stricter checks), multicollinearity is present.
  - Examine condition number of the design matrix (via numpy.linalg.cond); >30 suggests instability.

These diagnostics would be run on a training split (e.g., 80/20 train-test) to avoid data leakage.

#### Step 2: Address Heteroscedasticity
Once confirmed, prioritize fixes that preserve interpretability while stabilizing variance.
- **Variable Transformations**:
  - Apply log-transformation to the target (price) and/or continuous predictors (area, rooms) to stabilize variance, as real estate prices are often right-skewed and multiplicative. For example: `log_price = np.log(price + 1)` (add 1 to handle zeros). This is simple and effective for heteroscedasticity driven by scale.
  - If location contributes (e.g., urban vs. rural variance), include interaction terms like `area * location_dummy` to capture varying relationships.
- **Robust Estimation Techniques**:
  - Switch to weighted least squares (WLS) if you can estimate weights (e.g., inverse of absolute residuals from OLS). In statsmodels: `sm.WLS(y, X, weights=1/np.abs(residuals))`.
  - Use heteroscedasticity-consistent standard errors (HCSE, e.g., White's covariance estimator) in OLS. This doesn't change coefficients but adjusts t-stats and p-values for valid inference: `sm.OLS(y, X).fit(cov_type='HC3')`.
  - If severe, consider generalized least squares (GLS) assuming a specific variance structure (e.g., variance proportional to area).
- **Model Respecification**:
  - Add omitted variables (e.g., age of house, amenities) or polynomial terms (e.g., area²) if misspecification causes the issue. Use stepwise selection or LASSO to identify them.
  - Evaluate post-fix: Re-run residuals plot and Breusch-Pagan test; aim for random scatter around zero with constant spread.

#### Step 3: Address Multicollinearity
Focus on reducing redundancy without losing key information.
- **Variable Selection and Removal**:
  - Drop one of the highly correlated variables if domain knowledge allows (e.g., keep area over rooms if area is more predictive for prices). Use correlation matrix and VIF to prioritize.
  - For location (categorical), use one-hot encoding but drop a reference category to avoid perfect multicollinearity.
- **Dimensionality Reduction**:
  - Apply Principal Component Analysis (PCA) on correlated features (e.g., combine area and rooms into a "size" component). Fit PCA on standardized predictors, then regress price on the principal components. This decorrelates inputs but sacrifices interpretability—use only if coefficients aren't needed for business insights.
- **Regularization Methods**:
  - Switch to Ridge regression (L2 penalty) to shrink coefficients of correlated variables: `Ridge(alpha=1.0)` in scikit-learn. Tune alpha via cross-validation.
  - Or use Lasso (L1 penalty) for feature selection: It can set some coefficients to zero. Elastic Net combines both for better handling of grouped correlations.
  - Re-check VIF after; regularization often reduces it implicitly.
- **Data Collection**:
  - If feasible, gather more data (e.g., expand dataset to include more diverse listings) to improve the design matrix's stability. In real estate, this could mean scraping additional MLS data or partnering for more listings.

#### Step 4: Ensure Overall Model Robustness
After addressing the issues, validate and refine the model holistically.
- **Feature Engineering for Real Estate Specifics**:
  - Enhance location: Use geospatial features (e.g., distance to city center via haversine formula) or clustering (K-means on coordinates) instead of simple dummies to reduce multicollinearity from many categories.
  - Handle outliers: Real estate has extremes (e.g., luxury mansions); use robust scalers (e.g., RobustScaler) or winsorize at 1-99 percentiles.
  - Interactions and non-linearity: Add terms like `rooms * area` or use splines (e.g., via patsy in Python) if linear assumptions fail.
- **Model Validation and Selection**:
  - Use k-fold cross-validation (e.g., 5-fold) to assess performance on hold-out data. Metrics: R² for fit, RMSE/MAE for prediction error (focus on RMSE for price scale), and mean absolute percentage error (MAPE) for interpretability.
  - Compare models: After fixes, benchmark against alternatives like Random Forest or Gradient Boosting (e.g., XGBoost), which handle heteroscedasticity and multicollinearity inherently via ensemble methods. If linear remains superior for interpretability, stick with it.
  - Check for overfitting: Plot learning curves (train vs. validation error); if gap is large, add more regularization.
- **Inference and Deployment Considerations**:
  - Ensure coefficients are interpretable (e.g., "a 100 sq ft increase raises price by $X, holding others constant").
  - Test for other assumptions: Normality of residuals (Q-Q plot, Shapiro-Wilk test) and no autocorrelation (Durbin-Watson test, especially if time-series elements like market trends).
  - Deploy with monitoring: In production, track prediction errors on new data and retrain quarterly, as real estate markets fluctuate (e.g., due to interest rates).
  - Ethical checks: Audit for bias (e.g., location proxies for socioeconomic factors) to avoid discriminatory pricing.

By following these steps, the model becomes more reliable, with stable coefficients, valid p-values, and better generalization.
